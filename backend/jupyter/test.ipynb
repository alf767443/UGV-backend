```python
# Position database
from pymongo import MongoClient
import bson, math, datetime
from random import random, randint

size = 10000
client = MongoClient('mongodb://localhost:27017/')['CeDRI_UGV']['PositionApplication_position']
result = client.delete_many({})
try:
    print(result)
except:
    print('Fail')

dateTime = [
    {
        "dateTime": datetime.datetime.now()
    },
    {
        "dateTime": datetime.datetime.now()
    },
    {
        "dateTime": datetime.datetime.now()
    }
]

Odometry = [
    {
        "Left": 0.0,
        "Right": 0.0,
    },
    {
        "Left": 0.0,
        "Right": 0.0,
    },
    {
        "Left": round(random(),3) - round(random(),3),
        "Right": round(random(),3) - round(random(),3)
    }
]

Velocity = [
    {
        "Left": 0.0,
        "Right": 0.0,
    },
    {
        "Left": (Odometry[1]["Left"] - Odometry[0]["Left"]) / (dateTime[1]["dateTime"] - dateTime[0]["dateTime"]).total_seconds() if (dateTime[1]["dateTime"] - dateTime[0]["dateTime"]).total_seconds() != 0 else 0,
        "Right": (Odometry[1]["Right"] - Odometry[0]["Right"]) / (dateTime[1]["dateTime"] - dateTime[0]["dateTime"]).total_seconds() if (dateTime[1]["dateTime"] - dateTime[0]["dateTime"]).total_seconds() != 0 else 0
    },
    {
        "Left": (Odometry[2]["Left"] - Odometry[1]["Left"]) / (dateTime[2]["dateTime"] - dateTime[1]["dateTime"]).total_seconds() if (dateTime[2]["dateTime"] - dateTime[1]["dateTime"]).total_seconds() != 0 else 0,
        "Right": (Odometry[2]["Right"] - Odometry[1]["Right"]) / (dateTime[2]["dateTime"] - dateTime[1]["dateTime"]).total_seconds() if (dateTime[2]["dateTime"] - dateTime[1]["dateTime"]).total_seconds() != 0 else 0
    }
]

Acceleration = [
    {
        "Left": 0.0,
        "Right": 0.0,
    },
    {
        "Left": 0.0,
        "Right": 0.0,
    },
    {
        "Left": (Velocity[2]["Left"] - Velocity[1]["Left"]) / (dateTime[2]["dateTime"] - dateTime[0]["dateTime"]).total_seconds() if (dateTime[2]["dateTime"] - dateTime[0]["dateTime"]).total_seconds() != 0 else 0,
        "Right": (Velocity[2]["Right"] - Velocity[1]["Right"]) / (dateTime[2]["dateTime"] - dateTime[0]["dateTime"]).total_seconds() if (dateTime[2]["dateTime"] - dateTime[0]["dateTime"]).total_seconds() != 0 else 0
    }
]

def NewPosition(dateTime, Odometry, Velocity, Acceleration):
    Odometry.pop(0)
    Odometry.append(
        {
            "Left": round(random()- random(),3),
            "Right": round(random() - random(),3)
        }
    )
    
    Velocity.pop(0)
    Velocity.append(
        {
            "Left": round((Odometry[2]["Left"] - Odometry[1]["Left"]) / (dateTime[2]["dateTime"] - dateTime[1]["dateTime"]).total_seconds() if (dateTime[2]["dateTime"] - dateTime[1]["dateTime"]).total_seconds() != 0 else 0,3),
            "Right": round((Odometry[2]["Right"] - Odometry[1]["Right"]) / (dateTime[2]["dateTime"] - dateTime[1]["dateTime"]).total_seconds() if (dateTime[2]["dateTime"] - dateTime[1]["dateTime"]).total_seconds() != 0 else 0,3)
        }
    )

    Acceleration.pop(0)
    Acceleration.append(
        {
            "Left": round((Velocity[2]["Left"] - Velocity[1]["Left"]) / (dateTime[2]["dateTime"] - dateTime[1]["dateTime"]).total_seconds() if (dateTime[2]["dateTime"] - dateTime[1]["dateTime"]).total_seconds() != 0 else 0,3),
            "Right": round((Velocity[2]["Right"] - Velocity[1]["Right"]) / (dateTime[2]["dateTime"] - dateTime[1]["dateTime"]).total_seconds() if (dateTime[2]["dateTime"] - dateTime[1]["dateTime"]).total_seconds() != 0 else 0,3)
        }
    )

    result = {
        "dateTime": dateTime[2]['dateTime'],
        "Odometry": Odometry[2],
        "Velocity": Velocity[2],
        "Acceleration": Acceleration[2],
        "Fiducial":{
            "Mark": chr(randint(65, 70))
        },
        "Gyroscope":{
            "Angle": round(random()*(math.pi*2),3)
        },
        "GlobalPosition":{
            "X": round(random()*100,3),
            "Y": round(random()*100,3)
        }
    }

    return result

for j in range(0,size):  

    dateTime.pop(0)
    dateTime.append(
        {
            "dateTime": dateTime[1]["dateTime"] + datetime.timedelta(seconds=random()*10)
        }
    )
      
    #print(_bson)
    try:
        result = client.insert_one(NewPosition(dateTime=dateTime, Odometry=Odometry, Velocity=Velocity, Acceleration=Acceleration))
        #print(_bson)
    except:
        print('Fail')
```

```python
# Motor database
from pymongo import MongoClient
import bson, math, datetime
from random import random, randint

size = 10000
client = MongoClient('mongodb://localhost:27017/')['CeDRI_UGV']['MotorApplication_motor']
result = client.delete_many({})
try:
    print(result)
except:
    print('Fail')

dateTime = [
    {
        "dateTime": datetime.datetime.now()
    },
    {
        "dateTime": datetime.datetime.now()
    },
    {
        "dateTime": datetime.datetime.now()
    }
]

Voltage = [
    {
        "Left": 0.0,
        "Right": 0.0,
    }
]

Current = [
    {
        "Left": 0.0,
        "Right": 0.0,
    }
]

PWM = [
    {
        "Left": 0.0,
        "Right": 0.0,
    }
]

Power = [
    {
        "Left": 0.0,
        "Right": 0.0,
    }
]

def NewMotor(dateTime, Voltage, Current, PWM, Power, BatteryVoltage):
    PWM.pop(0)
    PWM.append(
        {
            "Left": random(),
            "Right": random()
        }
    )

    Current.pop(0)
    Current.append(
        {
            "Left": PWM[0]["Left"]*random()*10,
            "Right":PWM[0]["Right"]*random()*10 
        }
    )

    Voltage.pop(0)
    Voltage.append(
        {
            "Left": PWM[0]["Left"]*BatteryVoltage,
            "Right":PWM[0]["Right"]*BatteryVoltage 
        }
    )

    Power.pop(0)
    Power.append(
        {
            "Left": Voltage[0]["Left"]*Current[0]["Left"],
            "Right":Voltage[0]["Right"]*Current[0]["Right"]
        }
    )

    result = {
        "dateTime": dateTime[2],
        "PWM": PWM[0],
        "Current": Current[0],
        "Voltage": Voltage[0],
        "Power": Power[0]
    }

    return result

for j in range(0,size):  

    dateTime.pop(0)
    dateTime.append(
        {
            "dateTime": dateTime[1]["dateTime"] + datetime.timedelta(seconds=random()*10)
        }
    )
      
    #print(_bson)
    try:
        result = client.insert_one(NewMotor(dateTime=dateTime, Voltage=Voltage, Current=Current, PWM=PWM, Power=Power, BatteryVoltage=10))
        #print(_bson)
    except:
        print('Fail')
```
